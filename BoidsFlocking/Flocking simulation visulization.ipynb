```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
```

```python

# Load the corrected data
metric_df = pd.read_csv("Flocking_10run_metrics.csv")

# Prepare bar plot data
runs = list(range(1, len(metric_df)+1))
evacuated = metric_df["num_evacuated"]
burned = metric_df["num_burned"]

# Plot
plt.figure()
bar_width = 0.25
index = range(len(runs))

plt.bar(index, evacuated, bar_width, label='Evacuated', color='green')
plt.bar([i + bar_width for i in index], burned, bar_width, label='Burned', color='red')

plt.xlabel('Simulation Run')
plt.ylabel('Number of Agents')
plt.title('Total Evacuated vs Burned Per Run (Boids Flocking)')
plt.xticks([i + bar_width/2 for i in index], runs)
plt.legend()
plt.tight_layout()
plt.grid(True, alpha=0.5)
plt.show()
```

```python
# average escape times for each actor type
actor_types = ["staff_avg_escape_time", "adult_avg_escape_time",
               "patient_avg_escape_time", "child_avg_escape_time"]

plt.figure(figsize=(8,6))
for col in actor_types:
    sns.lineplot(data=metric_df, x=metric_df.index, y=col, label=col.replace("_avg_escape_time", "").capitalize())

plt.title("Average escape time for each actor")
plt.xlabel("Run Number")
plt.ylabel("Escape time (seconds)")
plt.legend(title="Actor Type")
plt.grid(True, alpha=0.5)
plt.show()
```
