# Import libaries

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

# Combine data

```python
df1 = pd.read_csv('advanced_evacuation_results_for_greedy_1_exits.csv')
df2 = pd.read_csv('advanced_evacuation_results_for_greedy_4_exits.csv')

df = pd.concat([df1, df2], ignore_index=True)
print(df)

# df.to_csv('combined_results.csv', index=False)
```

# Grouped bar chart for average escape time

```python
#postion and width
bar_width = 0.2
x = np.arange(len(df['n_exits']))

#grouped bar chart
plt.figure(figsize=(10, 6))
plt.bar(x - 1.5*bar_width, df['staff_avg_escape_time'], width=bar_width, label='Staff')
plt.bar(x - 0.5*bar_width, df['adult_avg_escape_time'], width=bar_width, label='Adult')
plt.bar(x + 0.5*bar_width, df['patient_avg_escape_time'], width=bar_width, label='Patient')
plt.bar(x + 1.5*bar_width, df['child_avg_escape_time'], width=bar_width, label='Child')

#formating and lables 
plt.xlabel('Number of Exits')
plt.ylabel('Average Escape Time')
plt.title('Average Escape Time by Actor Type vs Number of Exits')
plt.xticks(x, df['n_exits'])
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.show()
plt.savefig("evacuation_times_by_exits.png", dpi=300)
```
